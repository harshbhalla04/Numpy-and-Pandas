### Concatination of series and array

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
ser1 = pd.Series(np.random.randint(10,30,10))
ser2 = pd.Series(np.random.randint(10,30,10))
ser3 = pd.Series(np.random.randint(10,30,10))
```

```python
ser1
```

```python
# Concatenate ser1 and ser2
s=pd.concat([ser1,ser2])
s
```

```python
s[3]
```

```python
s=pd.concat([ser1,ser2],ignore_index=True)
s
```

```python
s[1]
```

```python
# concatenate ser1 and ser2 with hierarchial indexing
s=pd.concat([ser1,ser2],keys=['CLASS-A','CLASS-B'])
s
```

```python
s['CLASS-B'][1]
```

```python
# Concatenate the series with regular indexes (ignoring origional indexes)
pd.concat([ser1,ser2,ser3],ignore_index=True)
```

```python
s=pd.concat([ser1,ser2,ser3],axis=1,keys=['CLASS-A','CLASS-B','CLASS-C'])     # column wise operation
s
```

```python
s['CLASS-A'][2]
```

### Joining the dataframe

```python
# perform left join on df1 and df3 (default) on 'Emp ID'
df1
```

```python
df3
```

```python
df1.set_index('Emp ID',inplace=True)
```

```python
df3.set_index('Emp ID',inplace=True)
```

```python
df1.join(df3,lsuffix='_L',rsuffix='_R')
```

```python
# perform right join on df1 and df3 on 'Emp ID'
df1.join(df3,lsuffix='_L',rsuffix='_R',how='right')
```

```python
# perform inner join on df1 and df3 on 'Emp ID'
df1.join(df3,lsuffix='_L',rsuffix='_R',how='inner')
```

```python
# perform outer join on df1 and df3 on 'Emp ID'
df1.join(df3,lsuffix='_L',rsuffix='_R',how='outer')
```

### Merge the dataframe

```python
# Inner Join dataframe df1 and df3 using merge (default)
df1.merge(df3)                        # it first search for common column and make all comman column as key
```

```python
# merge is more advance function to join it automatically identify all common column as keys, also there is 
# no need to set key column as index by default merge perfrom inner join
```

```python
df1.merge(df3,on='Emp ID',suffixes=('_L','_R'),indicator=True)          # by default _x,_y
```

```python
# left Join dataframe df1 and df3 using merge
df1.merge(df3,on='Emp ID',suffixes=('_L','_R'),indicator=True,how='left')
```

```python
# Right Join dataframe df1 and df3 using merge
df1.merge(df3,on='Emp ID',suffixes=('_L','_R'),indicator=True,how='right')
```

```python
# Outer Join dataframe df1 and df2 using merge
df1.merge(df3,on='Emp ID',suffixes=('_L','_R'),indicator=True,how='outer')
```
