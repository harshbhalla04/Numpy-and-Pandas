### Ranking

```python
ser = pd.Series([23,78,None,90,45,63,82,None,55,49,78,None,34,21])
ser
```

```python
ser.rank()          # rankall the value, if we have same value by default it give average rankinhg to them, 'Dense' doesn't skip
                    # it give min value to same value and start with next number as rank, "first" it give min range to first 
                    # coming same value and next to the another same value 
```

```python
ser.rank(method='dense')
```

```python
ser.rank(method='dense',na_option='bottom',ascending=False)     # (na_option) --> top/bottom give rank to NaN resp.
```
